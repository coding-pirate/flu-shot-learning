```python
import pandas as pd
```

```python
#Reading our CSV file as a Pandas dataframe.
df = pd.read_csv('training_set_features.csv')
```

```python
#This function takes in a dataframe (X), 
#then replaces some of its ordinal labeled data with numbers
#and returns an updated dataframe.
def label_encoding(X):
        age_map = {
        '18 - 34 Years':1,
        '35 - 44 Years':2,
        '45 - 54 Years':3,
        '55 - 64 Years':4,
        '65+ Years':5
        }

        education_map = {
            '< 12 Years':1,
            '12 Years':2,
            'Some College':3,
            'College Graduate':4
        }

        income_map = {
            'Below Poverty':1,
            '<= $75,000, Above Poverty':2,
            '> $75,000':3
        }

        X['age_group_numerical'] = X.age_group.map(age_map)
        X['education_numerical'] = X.education.map(education_map)
        X['income_poverty_numerical'] = X.income_poverty.map(income_map)
        return X.drop(columns=['age_group','education','income_poverty'])
```

```python
#This function takes in a dataframe (X),
#then calls the label_encoding function on it,
#then calls the get_dummies function on it,
#and returns an updated dataframe (Y).
def dummy_encoding(X):
    Y = pd.get_dummies(label_encoding(X), prefix_sep='_', drop_first=True)
    return Y
```

```python
#Calling the dummy_encoding function on the dataframe we imported (df), 
#and then exporting the resulting dataframe as a CSV file.
dummy_encoding(df).to_csv('encoded_training_set_features.csv')
```
